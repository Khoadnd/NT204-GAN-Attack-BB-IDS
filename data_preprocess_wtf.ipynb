```python
import pandas as pd
from sklearn.preprocessing import Normalizer
from const import *
```

```python
X_train = pd.read_csv('dataset/NSL-KDD/KDDTrain+.txt', header=None, names=col_names_org)
y_train = X_train.iloc[:,-2]
X_train = X_train.iloc[:,:-2]

X_test = pd.read_csv('dataset/NSL-KDD/KDDTest+.txt', header=None, names=col_names_org)
y_test = X_test.iloc[:,-2]
X_test = X_test.iloc[:,:-2]

# only get content features

X_train = X_train[content_feature]
X_test = X_test[content_feature]
```

```python
display(X_train.describe(), X_test.describe())
```

```python
X_train_scale = X_train[normalize_these]
X_test_scale = X_test[normalize_these]
```

```python
scaler = Normalizer()
scaler.fit(X_train_scale)
X_train_scale = pd.DataFrame(scaler.transform(X_train_scale),columns=normalize_these)
X_test_scale = pd.DataFrame(scaler.transform(X_test_scale),columns=normalize_these)
```

```python
X_train[normalize_these] = X_train_scale
X_test[normalize_these] = X_test_scale
```

```python
display(X_train.describe(), X_test.describe())
```

```python
def get_key_from_value(value):
    if value == 'normal':
        return 0

    for key, vals in attack_category_dict.items():
        if value in vals:
            return key

y_train = y_train.apply(get_key_from_value)
y_test = y_test.apply(get_key_from_value)
```

```python
X_train = X_train.join(y_train)
X_test = X_test.join(y_test)

X_train.to_csv('dataset/train_content_feature.csv', index=False)
X_test.to_csv('dataset/test_content_feature.csv', index=False)
```
